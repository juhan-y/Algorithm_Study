### 오답

```python
n = int(input())

for i in range(n):
    vps = input()
    
    if vps.count('(') == vps.count(')'):
        print('YES')
    else:
        print('NO')
```

### 정답

```python
n = int(input())

for _ in range(n):
    vps = input()
    stack = []
    
    for i in vps:
        if i == '(':
            stack.append(i)
        
        if i == ')':
            if '(' in stack:
                stack.pop()
            else:
                stack.append(i)
                
    if len(stack) == 0:
        print('YES')
        
    else:
        print('NO')
```
